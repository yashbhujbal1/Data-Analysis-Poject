### 1. Google play store apps and review

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
duplicate_apps = pd.read_csv("apps.csv")
duplicate_apps
```

```python
apps = duplicate_apps.drop_duplicates()
apps
```

```python
print("Total number of apps in the dataset = ",len(apps))
```

```python
n=5
apps.sample(n)
```

### 2. Data cleaning

```python
# lsit of character to remove
chars_to_remove = ['+',',','$']

#list of column names to clean
cols_to_remove = ['Installs','Price']

for col in cols_to_remove:
    for char in chars_to_remove:
        apps[col] = apps[col].apply(lambda x : x.replace(char, ''))

apps
```

```python
apps.info()
```

### 3. Correcting data types

```python
apps['Installs'] = apps['Installs'].astype('float64')
apps['Price'] = apps['Price'].astype('float64')

print(apps.dtypes)
```

### 4. Exploring app categories

```python
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.graph_objs as go
```

```python
# print total number of unique chategories

num_categories = len(apps['Category'].unique())
print("Number of Categories = ",num_categories)
```

```python
# count the number of apps in each Category

num_apps_in_category = apps['Category'].value_counts()

#sort num_apps_in_category in descending order
sorted_num_apps_in_category = num_apps_in_category.sort_values(ascending = False)
sorted_num_apps_in_category
```

```python
sns.set_theme(style="whitegrid")
```

```python
fig  = plt.figure(figsize=(15,15))
x = num_apps_in_category.index
y = num_apps_in_category.values
plt.barh(x,y)
plt.xlabel("Number of apps")
plt.ylabel("Category")
plt.title("Number of apps in each Category")
plt.show()
```

```python
data = [go.Bar(x = num_apps_in_category.index,y = num_apps_in_category.values)]
plotly.offline.iplot({'data':data}) 
```

### 5. Distribution of apps ratings

```python
# Average rating of apps
avg_app_rating = apps['Rating'].mean()
print("Average app rating = ",avg_app_rating)
```

```python
# Distribution of apps according to their ratings
data = [go.Histogram(x = apps['Rating'])]

#vertical dashed line to indicate the average app ratings
layout  = {'shapes': [{'type':'line', 
                       'x0':avg_app_rating,
                       'y0':0,
                       'x1':avg_app_rating,
                       'y1':1000,
                       'line':{'dash':'dashdot'}
                      }]
          }

plotly.offline.iplot({'data':data, 'layout':layout})
```

### 6. Size and price of an app

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Select rows where both 'Rating' and 'Size' values are present (i.e the two values are not null)
apps_with_size_and_rating_present = apps.dropna()
```

```python
# Subset for categories with atleast 250 apps
large_categories = apps_with_size_and_rating_present.groupby('Category').filter(lambda x : len(x) >= 250)
large_categories
```

```python
# Plot size vs. rating
plt1  = sns.jointplot(x = large_categories['Size'], y = large_categories['Rating'])
```

```python
# Select apps whose 'Type' is 'Paid'
paid_apps = apps_with_size_and_rating_present[apps_with_size_and_rating_present['Type'] == 'Paid']
paid_apps
```

```python
# plot price vs. rating
plt2 = sns.jointplot(x = paid_apps['Price'], y = paid_apps['Rating'])
```

### 7. Relation between app category and app price

```python
fig, ax = plt.subplots()
fig.set_size_inches(15,8)

# Select a few popular categories
popular_app_cats = apps[apps.Category.isin(['GAME','FAMILY','PHOTOGRAPHY','MEDICAL','TOOLS','FINANCE','LIFESTYLE','BUSINESS'])]

# Examine the price trend by plotting price vs Category
ax = sns.stripplot(x = popular_app_cats['Price'], y = popular_app_cats['Category'], jitter=True, linewidth=1)
ax.set_title('App pricing trend across categories')
```

```python
# Apps whose price is greater than 200
apps_above_200 = popular_app_cats[['Category','App','Price']][popular_app_cats['Price'] > 200]
apps_above_200
```

### 8. Filter out "Junk" apps

```python
# Select app priced below $100
apps_under_100 = popular_app_cats[popular_app_cats['Price'] < 100]

fig, ax = plt.subplots()
fig.set_size_inches(15,8)

#Examine price vs category with the authentic apps(apps under 100)
ax = sns.stripplot(x = 'Price', y = 'Category', data=apps_under_100, jitter=True, linewidth=1)
ax.set_title('App pricing trend across categories after filtering for junk apps')
```

### 9. Popularity of paid apps vs free apps

```python
trace0 = go.Box(y = apps[apps['Type'] == 'Paid']['Installs'], name = 'Paid')  #data for paid apps

trace1 = go.Box(y = apps[apps['Type'] == 'Free']['Installs'], name = 'Free')  #data for free apps

layout = go.Layout(title = "Number of dowloads of paid apps vs free apps",
                   yaxis = dict(title = "Log number of downloads", type='log', autorange = True))

data = [trace0, trace1]
plotly.offline.iplot({'data':data, 'layout':layout})
```

### 10. Sentiment analysis of user reviews

```python
reviews_df = pd.read_csv("user_reviews.csv")
reviews_df
```

```python
# Join the two dataframes 
merged_df = pd.merge(apps, reviews_df, on='App', how="inner")
merged_df
```

```python
# Drop NA values from Sentiment and Reviews columns
merged_df = merged_df.dropna(subset=['Sentiment', 'Translated_Review'])
merged_df
```

```python
sns.set_style('ticks')
fig, ax = plt.subplots()
fig.set_size_inches(11,8)

# User reviews sentiment polarity for paid vs. free apps
ax = sns.boxplot(x = 'Sentiment_Polarity', y = 'Type', data = merged_df)
ax.set_title('Sentiment Polarity Distribution')
```
